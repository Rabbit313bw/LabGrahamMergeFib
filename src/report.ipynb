```python
import time
import tqdm
time_merged = []
time_heap = []
for n in tqdm.tqdm(range(10, 10000, 10)):
    data = generate_random_data(n, 1, 10)
    time_start = time.time()
    MSP_merged = grahamscan(data)
    time_finish = time.time()
    time_merged.append(time_finish - time_start)
    time_start = time.time()
    MSP_heap = grahamscan(data, sort=heap_sort)
    time_finish = time.time()
    time_heap.append(time_finish - time_start)
```

```python
len(time_heap)
```

```python
plt.plot(list(range(10, 10000, 10)), time_merged, label="merged_sort")
plt.plot(list(range(10, 10000, 10)), time_heap, label="heap_sort_with_fib_heap")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def nlogn(n: float, c: float = 1.0) -> float:
    return c * n * np.log2(n)
plt.plot(range(1, 10000, 1), [nlogn(n) for n in range(1, 10000, 1)])
plt.plot(range(1, 10000, 1), [nlogn(n, 5) for n in range(1, 10000, 1)])
```
